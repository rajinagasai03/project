```python
#check the skewness of the attributes
import numpy as np
intFloatboDf=boDf.select_dtypes([np.int,np.float])
for i, col in enumerate(intFloatboDf.columns):
    if (intFloatboDf[col].skew()>= 0.1):
        print("\n",col,"feature is positively skewed and value is:", intFloatboDf[col].skew()) 
    elif (intFloatboDf[col].skew()<= -0.1):
        print("\n",col,"feature is negtively skewed and value is:", intFloatboDf[col].skew())
    else:
        print("\n",col,"feature is normally distributed and value is:", intFloatboDf[col].skew())
```

```python
import pandas as pd
import numpy as np


# Load the dataset
df = pd.read_csv('c:\cc\ebd.csv')

# Convert relevant columns to numerical data types
numerical_columns = ['Base Salary', 'Overtime', 'Total Compensation']

# Remove any non-numeric characters (e.g., $, commas) and convert to float
df[numerical_columns] = df[numerical_columns].replace({'\$': '', ',': ''}, regex=True).astype(float)

# Calculate the skewness of each numerical column
skewness_results = df[numerical_columns].skew()

# Output the skewness results
print("Skewness of numerical columns:")
print(skewness_results)
```
