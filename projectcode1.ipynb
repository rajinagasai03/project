```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have the payroll data in a pandas DataFrame named 'payroll_data'
# If needed, convert the numerical columns from object to numeric type
numerical_columns = ['REGULAR', 'RETRO', 'OTHER', 'OVERTIME', 'INJURED', 'DETAIL', 'QUINN_EDUCATION']
payroll_data[numerical_columns] = payroll_data[numerical_columns].apply(pd.to_numeric, errors='coerce')

# Create a correlation matrix for the numerical columns
correlation_matrix = payroll_data[numerical_columns].corr()

# Create a heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', annot_kws={'size': 10})
plt.title("Correlation Heat Map for Payroll Data")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```
